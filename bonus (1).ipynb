```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameters

alpha = 3.0
L = 100
W = 100
T_end = 100  # Total time
dx = 1.0
dy = 1.0
dt = 0.1  # Time step

nx = int(L/dx) + 1
ny = int(W/dy) + 1
nt = int(T_end/dt)

```

```python
# Creating grid
x = np.linspace(0, L, nx)
y = np.linspace(0, W, ny)
u = np.zeros((nx, ny))

# Initial condition
u[:, :] = 0

# Boundary conditions
u[:, 0] = 200
u[:, -1] = 200
u[0, :] = 0
u[-1, :] = 0
```

```python
# Time-stepping loop
for t in range(nt):
    un = u.copy()
    for i in range(1, nx-1):
        for j in range(1, ny-1):
            u[i, j] = (un[i, j] + alpha * dt * (
                       (un[i+1, j] - 2*un[i, j] + un[i-1, j]) / dx**2 +
                       (un[i, j+1] - 2*un[i, j] + un[i, j-1]) / dy**2))
```

```python
# Plot the final solution
X, Y = np.meshgrid(x, y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, u.T, cmap='hot')
plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('Temperature')
plt.title('Temperature Distribution at Final Time')
plt.show()
```
